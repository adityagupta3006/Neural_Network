```python
import numpy as np

import matplotlib
matplotlib.use('TkAgg')
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return (1/(1+ np.exp(-x)))
```

```python
def calc_z(theta, j):
    return theta.dot(j)
```

```python
def calc_alpha(z):
    alpha = []
    for m in z:
        alpha.append(sigmoid(m))
    return alpha
```

```python
def calc_total_error(a):
    total_error = (pow((y[0]-a[0]), 2))*0.5 + (pow((y[1]-a[1]), 2))*0.5 
    return total_error
```

```python
# a is aplha out and a_h is alpha hidden
def calc_theta_new(theta_2, a, a_h):
    
    # calculating theta 2
    term_0 = (a[0]-y[0])*a[0]*(1-a[0])
    term_1 = (a[1]-y[1])*a[1]*(1-a[1])
    
    theta10 = term_0*a_h[0]
    theta11 = term_0*a_h[1]
    theta12 = term_0*a_h[2]
    theta20 = term_1*a_h[0]
    theta21 = term_1*a_h[1]
    theta22 = term_1*a_h[2]
    a = np.asarray(a)
    theta_2_new_dash = np.asarray([[theta10, theta11, theta12], [theta20, theta21, theta22]])
    theta_2_new = theta_2  - (a.dot(theta_2_new_dash))
    
    # calculating theta 1
    
    term_2 = (a_h[1]*(1-a_h[1]))
    term_3 = (a_h[2]*(1-a_h[2]))
    
    theta10_1 = (term_0*theta_2[0,1] + term_1*theta_2[1,1])*(term_2*x[0])
    theta11_1 = (term_0*theta_2[0,1] + term_1*theta_2[1,1])*(term_2*x[1])
    theta12_1 = (term_0*theta_2[0,1] + term_1*theta_2[1,1])*(term_2*x[2])
    theta20_1 = (term_0*theta_2[0,2] + term_1*theta_2[1,2])*(term_3*x[0])
    theta21_1 = (term_0*theta_2[0,2] + term_1*theta_2[1,2])*(term_3*x[1])
    theta22_1 = (term_0*theta_2[0,2] + term_1*theta_2[1,2])*(term_3*x[2])
    
    theta_1_new = np.asarray([[theta10_1, theta11_1, theta12_1], [theta20_1, theta21_1, theta22_1]])
    
    return theta_1_new, theta_2_new
```

```python
# defining initial and given values.
x = np.array([1, 0.05, 1])                  # first element is the bias.
y = np.array([0.01, 0.99])
iter = 300
theta_1 = np.array([[np.random.uniform(0,1) for m in range(3)] for n in range(2)])
theta_2 = np.array([[np.random.uniform(0,1) for m in range(3)] for n in range(2)])
print 'theta_1 is : \n', theta_1, '\n'
print 'theta_2 is : \n', theta_2

theta_1_array = []
theta_2_array = []
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.set_title('Neural Networks')
ax.set_xlabel('iteration')
ax.set_ylabel('cost')

for i in range(iter):
    # forward propogation and backward propogation.
    z1 = calc_z(theta_1, x)
    
    alpha_hidden_dash = calc_alpha(z1)
    
    alpha_hidden = [1] + alpha_hidden_dash

    z2 = calc_z(theta_2, alpha_hidden)

    alpha_out = calc_alpha(z2)
    
    total_error = calc_total_error(alpha_out)
    
    print "Total error for iteration ", i," is : ", total_error, "\n"

    theta_1, theta_2 = calc_theta_new(theta_2, alpha_out, alpha_hidden)
    theta_1_array.append(theta_1)
    theta_2_array.append(theta_2)
       
    ax.plot(i, total_error, linestyle='None', marker='o', markersize=5, color='blue', label='')
```
